En el gráfico ACF se evidencia que la serie tiene tendencia puesto quelos $\hat{\rho}(k)$ decaen muy lentamente, esto es un indicio de que hay que aplicar por lo menos una diferencia a la serie.

```python
plot_pacf(ts3, method= "ywm")
plt.show
```

El patrón de la gráfica parece ser decaimiento exponencial pero antes como se encontraron evidencias gráficas que suguieren que que el proceso no es estacionario se realiza la prueba Dickey-Fuller para confirmar estas sospechas, dicho test tiene como hipótesis:

$\left\lbrace\begin{array}{c}H_0: \phi = 1  \\ H_1: \phi < 1\end{array}\right. $ equivalentemente
$\left\lbrace\begin{array}{c} H_0: \textrm{El proceso no es estacionario} \\ H_1:\textrm{El proceso es estacionario.} \end{array}\right. $

Un proceso AR(p) o ARMA(p,q) es estacionario si y solo si $\phi_p(B)$ tiene todas sus raices con modulo mayor estrictamente a uno con lo que si hay raiz unitaria el proceso no puede ser estacionario.

```python
adfuller(ts3.registro) #p-valor <0.05 
```

Con un valor-p mas grande que el nivel de significancia $\alpha$= 0.05 no hay evidencias muestrales para rechazara la hpótesis nula, por lo que según el test de Dickey-Fuller el proceso no es estacionario congruente con lo hallado en el gráfico de registro vs tiempo. Al diferenciar una vez se obtiene lo siguiente: 


```python
ts3.diff().dropna().plot()
```

```python
adfuller(ts3.diff().dropna().registro)
```

este proceso luce mas estacionario que el anterior por mucho; si se aplica el test Dickey-Fuller sobre la serie diferenciada, se obtiene con un valor-P menor que cualquier nivel de significancia que en efecto con d=1 la serie ya es estacionaria; las gráficas ACF y PACF de la serie diferenciada son:

```python
plot_acf(ts3.diff().dropna())
plt.show()
plot_pacf(ts3.diff().dropna(), method="ywm")
plt.show()
```

El patron que siguen los $\hat{\rho}(k)$  es mucho mas claro que el anterior, es inegable que la ACF tiene un patrón de decaimiento exponencial sinusoidal  esto sucede cuando el polinomio $\phi_{p}(B)$ tiene al menos una raiz compleja por lo que p debe ser mayor o igual a dos; en cuanto al patrón de la PACF no es muy claro si es tipo corte o es tipo decaimiento exponencial, no se puede ignorar que los valores de $\hat{\phi}_{kk}$ cambian de signo "ciclicamente" esto puede ser un indicio de que el verdadero patron de la PACF es decaimiento exponencial podriamos proponer mas particularmente un decaimiento exponencial sinusoidal, de manera similar que para el polinomio $\phi_{p}(B)$ el polinomio $ϴ_{q}(B)$ debe tener al menos una raiz compleja por lo que q debe ser mayor o gual a dos, por lo que un modelo candidato es un ARIMA(2,1,2). Por proponer otro modelo se puede pensar que el paton de la grafica PACF es corte con ultimo corte diferente de cero en k= 11 pues a medida que el rezago aumenta tambien lo hace la probabilidad de cometer error tipo 1 asi el segundo modelo candidato sera ARIMA(11,1,0)

```python
#ts3.index = pd.DatetimeIndex(ts3.index).to_period('D')
mod1 = ARIMA(ts3, order=(2,1,2)).fit()
mod2 = ARIMA(ts3, order=(11,1,0)).fit()
mod3 = ARIMA(ts3, order=(4,1,4)).fit() #que sucede si se aumenta p y q
```

```python
print(mod1.aic, mod2.aic, mod3.aic)
print(mod1.bic, mod2.bic, mod3.bic)
```

El modelo que proporciona el mejor BIC es el modelo ARIMA(4,1,4) primero se verifica si el el modelo ARIMA(0,1,1) es invertible para ello verifiquemos si las raices del polinomio $ϴ_1(B)$ tienen modulo mayor a uno.

```python
mod3.summary().tables[1]
```

$$X_t= \sum_{i=1}^4 \phi_i X_{t-i}+a_t +\sum_{j=1}^4 ϴ_j a_{t-j}$$
$$(1-\phi_1B-\phi_2B^2-\phi_3B^3-\phi_4B^4)X_t = (1-ϴ_1B-ϴ_2B^2-ϴ_3B^3-ϴ_4B^4)a_t $$
$$\phi_4(B)X_t= ϴ_4(B)a_t$$
$$ϴ_4(B)= 1-0.8557B-0.7122B^2-1.0591B^3-0.4851B^4$$
$$ \phi_4(B)= 1-0.0794B+0.0318B^2-0.0591B^3+0.0587B^4 $$
Donde X_t la serie ya diferenciada

```python
coeffma= [-0.4851, -1.0591, -0.7122,-0.8557, 1]
print(abs(np.roots(coeffma)))
print(abs(np.roots(coeffma))>1)
```

El modelo ARIMA(4,1,4) no es invertible; ahora se verifica si las raices del polinomio $\phi_4(B)$ tienen modulo mayor a uno, como lo son se tiene que el modelo es estacionario.

```python
coeffar= [0.0587,-0.0591,0.0318,-0.0794,1]
print(abs(np.roots(coeffar)))
print(abs(np.roots(coeffar))>1)
```


¿Existe autocorrelación en los residuales del modelo seleccionado? en la ACF vemos que no hay correlacion significativa entre los residuales de ningun orden coherente con los resultados de la grafica se ven los valores mas exactos de estas correlaciones y se aplica el test  Ljung-Box para los primeros 10 rezagos.

```python
plot_acf(mod3.resid)
plt.show()
sm.stats.acorr_ljungbox(mod3.resid, lags=10, return_df=True)
```

Para todos los rezagos con el valor p no se encuentran evidencias para rechazar la hipótesis nula que es:
$$\left\lbrace\begin{array}{c}H_0: \rho(1) = \rho(2)=...= \rho(10) = 0  \\ H_1: \rho(k) \neq 0~ \textrm{para algun k=1,...,10}\end{array}\right. $$
se concluye que no existen correlaciones entre los errores separados hasta 10 rezagos.

Continuando con la prueba de los supuestos del modelo se procede a verificar si los errores se distribuyen normal

```python
mod3.plot_diagnostics(figsize=(15, 12))
plt.show()
mod3.summary().tables[2]
```

Se tienen buenos resultados en el gráfico Q-Q plot ya que los cuantiles muestrales de los residuales se asemejan a los cuantiles teoricos de la distribución normal, y con el test de Jarque- Bera se obtiene un valor-P mayor a 0.05 con lo que se dice los errores del modelo se distribuyen normal; además no existen evidencias encontra del supuesto de que la media de los errores es cero y hay homocedasticidad de varianza, con esto en mente se pasa a realizar los pronosticos

```python
fore = mod3.get_forecast(5) 
fore_ci = fore.conf_int()
print(fore.predicted_mean)
print(fore_ci)
```

```python
plt.figure(figsize=(15,15))  
ts3.registro.plot()
fore.predicted_mean.plot()
fore_ci.iloc[:,0].plot(color="green")
fore_ci.iloc[:,1].plot(color="green")
plt.legend(["Real", "Predicho", "Intervalo de Confianza"])
```
